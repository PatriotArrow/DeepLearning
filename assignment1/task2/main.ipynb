```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(t):
    return 1/(1 + np.exp(-t))
```

```python
file_name = r"../datasets/blobs600.csv"
df = pd.read_csv(file_name)

X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values.reshape(-1, 1)
print(X.shape, Y.shape)



```

```python

def loss_function(y, y_cap):
    epsilon = np.finfo(float).eps
    return -(y * np.log (y_cap) + (1- y ) * np.log(1 - y_cap))

```

```python
#for calculating mean loss over dataset
#takes per-sample losses and averges them, this gives us single scalar J which is the training objective to minimize
def cost_function(Y, Y_cap):
    return np.mean(loss_function(Y,Y_cap))
```

```python
#hyperparameters
alpha = 0.01
w = np.array(X.shape[1] * [.5]).reshape(-1, 1)
b = 0.5

#stopping = False; J_running = 0; J_running_prev = 0; iteration = 0

#training loop
max_iters = 10000
#Tol is stopping threshold based on cost improvement
tol = 1e-10
#previous cost which is set to infinity so first iteration always continues 
j_prev = np.inf
iteration = 0 #loop counter
```

```python
while iteration < max_iters :
    #forward pass 
    #linear combination for all samples 
    Z = X @ w + b
    Y_cap = sigmoid(Z)
    J = cost_function(Y, Y_cap)
    #if change in cost is very small then stop training
    if abs(j_prev - J) < tol:
        break
```

```python
#gradient (backpropagation)

dZ = (Y_cap - Y)
dW = (X.T @ dZ) / X.shape[0]
dB = np.sum(dZ) / X.shape[0]

#gradient descent 
w = w - alpha*dW
b = b - alpha*dB

j_prev = J
iteration += 1
```

```python
#Predictions and accuracy
Y_prob = sigmoid (X @ w + b)
#converting probablities to binary predictions where threshold is 0.5
Y_pred = (Y_prob >= 0.5).astype(int)
#computing training accuracy
acc = np.mean(Y_pred == Y)

print("Ireations:", iterations)
print("Final cost J :", float(J))
print("Training accuarcy:", float(acc))
print("wshape", w.shape, "b:",float(b))
```
